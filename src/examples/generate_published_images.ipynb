# Generate Published Images
In the Paper "Foaming by Arburg Plastic Freeformingprocess: Optical characterization ofchanges in the foam structure" (put link here), four synthetic images generated with this algorithm are shown (Figure 6). 
This notebook shows how to recreate similaar images, based on the saved configuration file of these images. The image generation algorithm is non deterministic and uses a lot of randomness in size and position of the foam pores and tracks. (To make the algotihm deterministic, you would add some (de-)serialization, saving and loading code to the Ellipse class in order to exactely save and load the drawn pores and tacks.)


First, add some imports and set some paths.
If you want to use a config in a different directory, adapt this here.

```python
from pathlib import Path
import os
import sys

repo_root = Path(os.getcwd()).parent.parent
sys.path.append(str(repo_root))

from src.parameters import Parameters, UserConfig
from src import draw_routine

path_to_configs = repo_root / "configs" 
output_directory = str(repo_root / "output")
```

Now load the UserConfig from a config file

```python
config_file_6b = str(path_to_configs / "config_014.json")
user_config6a = UserConfig.load_from_file(filename=config_file_6b, 
                                        output_directory=output_directory)
print(user_config6a)
```

And load the Parameters from the same config file

```python
parameters6a = Parameters.load_from_file(filename=config_file_6b)
print(parameters6a)
```

Adapt the porosity margins a bit to prevent the image generation algorithm from iterating to a precise solution.
(Since there is a lot of randomness included in the image generation process, the porosities belonging to the same configuration may vary a bit).

```python
parameters6a.total_porosity_margin = 3.0
parameters6a.porosity_foam_pore_margin = 2.0
```

And now generate the image!

```python
draw_routine.generate_image(user_config6a, parameters6a)
```
